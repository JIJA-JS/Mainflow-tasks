```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('Diabetes.csv')
```

```python
df.head()
```

```python
filtered_df = df[df['Age'] > 50]
print("Filtered DataFrame:\n", filtered_df)
```

```python
filtered_df = df[df['Glucose'] > 140]
print("Filtered DataFrame:\n", filtered_df)
```

```python
#Handling missing values
df_filled = df.fillna(value={'Age': 50, 'Glucose': 148})
print("\nDataFrame with filled missing values:\n", df_filled)
```

```python
df_dropped = df.dropna()
print("\nDataFrame with dropped rows:\n", df_dropped)
```

```python
# Calculate summary statistics for the entire DataFrame
summary_stats = df.describe()

# Display the results
print("\nSummary statistics:\n", summary_stats)
```
