```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_csv("Car Price.csv")
```

```python
data.head(5)
```

```python
#scatter plot
plt.scatter(data['Year'], data['Brand'])
plt.title("scatter plot")
plt.xlabel('Year')
plt.ylabel('Brand')
plt.show()
```

```python
#line chart
plt.plot(data['Year'])
plt.plot(data['Brand'])
plt.title("line chart")
plt.xlabel('Year')
plt.ylabel('Brand')
plt.show()
```

```python
#bar chart
plt.bar(data['Year'],data['Brand'])
plt.title("bar chart")
plt.xlabel('Year')
plt.ylabel('Brand')
plt.show()
```

```python
#histogram
plt.hist(data['Selling_Price'])
plt.title('histogram')
plt.show()
```
